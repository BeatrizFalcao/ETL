
## **Extração de dados**



```python
import pandas as pd
import numpy as np
from pandas.io import gbq
```

```python
albuns = pd.read_csv('spotify-albums_data_2023.csv')
artistas = pd.read_csv('spotify_artist_data_2023.csv')
trilhas=pd.read_csv('spotify_tracks_data_2023.csv')
```

```python
colunas_necessarias=['track_name', 'track_id', 'track_number', 'artist_id', 'album_id','album_type', 'album_name']
```

```python
albuns = albuns[colunas_necessarias]
```

```python
artistas = artistas.drop(['genre_0', 'genre_1', 'genre_2', 'genre_3', 'genre_4', 'genre_5', 'genre_6'], axis=1)
```

```python
trilhas = trilhas.drop(['explicit'], axis=1)
```

## **Transformação**

```python
albuns_artistas  = pd.merge(albuns, artistas, how = 'inner', left_on = 'artist_id', right_on = 'id')
```

```python
albuns_artistas.head(5)
```

```python
albuns_artistas=albuns_artistas.drop(columns=['id'])
```

```python
albuns_artistas.head(5)
```

```python
albuns_artistas_trilhas  = pd.merge(albuns_artistas, trilhas, how = 'inner', left_on = 'track_id', right_on = 'id')
```

```python
albuns_artistas_trilhas.head()
```

```python
albuns_artistas_trilhas=albuns_artistas_trilhas.drop(columns=['id'])
```

```python
albuns_artistas_trilhas.count()
```

```python
albuns_artistas_trilhas.dtypes
```

```python
albuns_artistas_trilhas['track_name'] = albuns_artistas_trilhas['track_name'].astype(str, copy=True, errors='raise')
```

```python
mais_populares = albuns_artistas_trilhas[albuns_artistas_trilhas.artist_popularity >= 100]
```

```python
mais_populares.head()
```

```python
mais_populares.count()
```

## **Carregamento**

```python
mais_populares.to_csv('mais_populares.csv')
```
